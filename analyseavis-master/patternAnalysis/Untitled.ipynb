```python
import pandas as pd
import glob, os
import re
from textblob import TextBlob

from collections import Counter

def remove_whitespace(text): 
    return  " ".join(str(text).split()) 

dfPARIS = pd.read_csv("./csv/paris.csv", sep='$')
dfPARIS['text_lw'] = dfPARIS['text'].str.lower()
dfPARIS['text_lw'][:5].apply(lambda x: str(TextBlob(x).correct()))
dfPARIS['text_lw'].apply(remove_whitespace)

dfMARS = pd.read_csv("./csv/marseille.csv", sep='$')
dfMARS['text_lw'] = dfMARS['text'].str.lower()
dfMARS['text_lw'][:5].apply(lambda x: str(TextBlob(x).correct()))
dfMARS['text_lw'].apply(remove_whitespace)

dfLYON = pd.read_csv("./csv/lyon.csv", sep='$')
dfLYON['text_lw'] = dfLYON['text'].str.lower()
dfLYON['text_lw'][:5].apply(lambda x: str(TextBlob(x).correct()))
dfLYON['text_lw'].apply(remove_whitespace)


dfMTP = pd.read_csv("./csv/mtp.csv", sep='$')
dfMTP['text_lw'] = dfMTP['text'].str.lower()
dfMTP['text_lw'][:5].apply(lambda x: str(TextBlob(str(x)).correct()))
dfMTP['text_lw'].apply(remove_whitespace)


dfTLS = pd.read_csv("./csv/toulouse.csv", sep='$')
dfTLS['text_lw'] = dfTLS['text'].str.lower()
dfTLS['text_lw'][:5].apply(lambda x: str(TextBlob(str(x)).correct()))
dfTLS['text_lw'].apply(remove_whitespace)


dfDIJON = pd.read_csv("./csv/dijon.csv", sep='$')
dfDIJON['text_lw'] = dfDIJON['text'].str.lower()
dfDIJON['text_lw'][:5].apply(lambda x: str(TextBlob(str(x)).correct()))
dfDIJON['text_lw'].apply(remove_whitespace)


dfRENNES = pd.read_csv("./csv/rennes.csv", sep='$')
dfRENNES['text_lw'] = dfRENNES['text'].str.lower()
dfRENNES['text_lw'][:5].apply(lambda x: str(TextBlob(str(x)).correct()))
dfRENNES['text_lw'].apply(remove_whitespace)
```

```python
import nltk

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()


def getSString(word):
    for n in range(1, len(str(word))):
        substr_counter = Counter(str(word)[i: i+n] for i in range(len(str(word)) - n))
        phrase, count = substr_counter.most_common(1)[0]
        if count >=2 and len(str(phrase)) > 5: 
            print(str(phrase))

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(str(text))]

    
frames = [dfPARIS, dfMARS, dfLYON, dfMTP, dfTLS, dfDIJON, dfRENNES]
dfALLCITY = pd.concat(frames)

dfALLCITY['text_lw'].apply(lemmatize_text)
dfALLCITY['text_lw'].apply(getSString)

```
